```python
def plot_graph(df):
  plt=sns.histplot(df,kde=True)
  plt.set_title(df.name + 'Distribution')
  plt.set_xlabel(df.name + 'class')
  plt.set_ylabel('Count')
```

```python
plot_graph(df['Glucose'])
```

```python
df[df.Outcome == False]
```

```python
def plot_graph_1(df,column):
  plt=sns.histplot(df[df.Outcome == False][column.name],color='green',label='Non-diabetic')
  plt=sns.histplot(df[df.Outcome == True][column.name],color='red',label='diabetic')
  plt.set_title('Distribution of'+column.name)
  plt.set_xlabel(column.name)
  plt.set_ylabel('counts')
  plt.axes.legend(loc=0)
```

```python
plot_graph_1(df,df.Glucose)
```

```python
plot_graph(df['BloodPressure'])
```
